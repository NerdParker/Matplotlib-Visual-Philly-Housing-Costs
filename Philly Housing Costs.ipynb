```python
# packages to store and manipulate data
import numpy as np
import pandas as pd

# plotting tools
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Import Dataset
df = pd.read_csv('C:/Users/607791/Desktop/DS/Data Visualization/Week 6/Week_6_Homework_Data.csv')
df.head()
```

```python
sq_ft = df['Number of Square Feet per $1,000,000']
city = df['City']
y_pos = np.arange(len(city))

# size of plot
plt.figure(figsize=(12,14.5))

ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)    

ax.xaxis.set_ticks_position("top")
ax.get_xaxis().tick_top()    
ax.get_yaxis().tick_left()  

# square footage bars
plt.barh(y_pos, sq_ft, color=['thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','thistle','blue','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey'], zorder=2)

# add cities to y-axis
plt.yticks(y_pos, city)

# axis labels/title
#plt.ylabel('Cities', fontsize=14)
plt.title('Philadelphia in top 1/3 of Housing Square Footage per $1,000,000', weight='bold', fontsize=22)
#plt.xlabel('Square Footage per $1,000,000', fontsize=14)
 
# Show graphic
plt.show()
```
